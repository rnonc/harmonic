```python
import tensorflow as tf
import os,glob,csv,scipy
from scipy.io import wavfile
import numpy as np

from IPython import display
import matplotlib.pyplot as plt
from progressbar import progressbar
```

```python
os.chdir('C:/Users/Admin/Documents/Python Scripts/project voice/DATA/Macron Voice')

wav_file_name = 'Mes chers compatriotes des Antilles.wav'
sample_rate, wav_data = wavfile.read(wav_file_name, 'rb')
```

```python
N = 1000
cut = 1000
skipping = 10
training_wave = wav_data[:200000,0]
```

```python
display.Audio(training_wave,rate=sample_rate)
```

```python
L_origin = []

for i in progressbar(range(N,len(training_wave),skipping)):
    L_origin += [np.fft.fft(training_wave[i-N:i])]
cutting = np.array(L_origin,dtype = np.complex64)[:,:cut]
rest = np.array(L_origin,dtype = np.complex64)[:,cut:]
cutting = cutting/(np.abs(cutting.max()))
cutting = cutting.view(np.float32)

training = cutting
```
